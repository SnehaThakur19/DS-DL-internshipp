```python
import pandas as pd
df = pd.read_csv ('titanic.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder (sparse_output = False, dtype ='int')
```

```python
df[['male','female']] = ohe.fit_transform(df[['Sex']])
```

```python
df.head(2)
```

```python
df.dropna(subset=['Embarked'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelEnc = LabelEncoder ()
embarkedEncoded = labelEnc.fit_transform (df[['Embarked']])
df['embarkedEnc'] = embarkedEncoded
```

```python
df.head()
```

```python
df = df.drop(columns = ['Embarked', 'Sex'])
```

```python
df.head()
```

```python

from sklearn.impute import SimpleImputer
si = SimpleImputer (strategy = 'mean')
ageImpute = si.fit_transform (df[['Age']])
```


df['imputedAge'] = ageImpute

```python
df = df.drop(columns = ['Age'])
```

```python

df.to_csv ('cleanedData.csv')
```
